```python
#2
df = pd.read_csv('/content/wrangling/assignment/data/sharks.csv', low_memory = False)
df.head(10)
```

```python
#2.1
import numpy as np
df['Type'].unique()
df['Type'].value_counts()
```

```python
#2.2
type2 = df['Type']
type2 = type2.replace(['Sea Disaster', 'Watercraft', 'Boat', 'Boating', 'Boatomg'], 'Vehicular') #Simplify categories related to vehicles
type2 = type2.replace(['Invalid', 'Questionable', 'Unconfirmed', 'Unverified', 'Under investigation'], 'Missing Data') #combine questionable categories

df['Type'] = type2
df['Type'].value_counts()
```

```python
#3
url = 'http://www.vcsc.virginia.gov/pretrialdataproject/October%202017%20Cohort_Virginia%20Pretrial%20Data%20Project_Deidentified%20FINAL%20Update_10272021.csv'
df = pd.read_csv(url,low_memory=False) # Pulling data from website
df.head(5)
```

```python
#3.1
released = df['WhetherDefendantWasReleasedPretrial']
released.unique()
released = released.replace(9, np.nan) #remove incomplete data points
print(released.value_counts())
df['WhetherDefendantWasReleasedPretrial'] = released #sent edits back to data frame
```

```python
#4
import pandas as pd
import numpy as np
sentence = df['ImposedSentenceAllChargeInContactEvent']
sentence = pd.to_numeric(sentence, errors='coerce') #sentence becomes numeric type
print(sum(sentence.isnull()))
type = df['SentenceTypeAllChargesAtConvictionInContactEvent']
print(pd.crosstab((sentence.isnull()), type), '\n') #Cat 4 is where charges were dismissed

sentence = sentence.mask(type == 4, 0) #Replace length with 0 when type is 4
sentence = sentence.mask(type == 9, np.nan) # Replace length with np.nan when type is 9

print(sum(sentence.isnull())) # see how many are missing -- better
print(pd.crosstab((sentence.isnull()), type)) #confirm its fixed

df['ImposedSentenceAllChargeInContactEvent'] = sentence #reassign back to dataframr
```
